```python
train = pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/train.csv')
train.head()
```

```python
train.describe()
```

```python
train['EJ'].replace({'A': 0, 'B': 1}, inplace = True) # replace category with number
train['EJ']
```

```python
from sklearn.preprocessing import StandardScaler

X_train = train.iloc[:, 1:-1]
y_train = train.iloc[:, -1]

scaler = StandardScaler()
X_train_standardized = scaler.fit_transform(X_train)
X_train_standardized_df = pd.DataFrame(X_train_standardized, columns = X_train.columns)
X_train_standardized_df.describe()
```

```python
test = pd.read_csv('/kaggle/input/icr-identify-age-related-conditions/test.csv')
X_test = test.iloc[:, 1:]
X_test['EJ'].replace({'A': 0, 'B': 1}, inplace = True) # replace category with number
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train_standardized_df, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_proba = model.predict_proba(X_test)
y_pred_proba_df = pd.DataFrame(y_pred_proba)
out = pd.concat([test.iloc[:, 0], y_pred_proba_df], axis = 1)
out.columns = ['Id', 'class_0', 'class_1']
out.to_csv('submission.csv', index = False)
```
